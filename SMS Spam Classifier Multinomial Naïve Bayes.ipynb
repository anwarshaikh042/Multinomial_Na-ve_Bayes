# SMS Spam Classifier: Multinomial Naive Bayes
The notebook is divided into the following sections:

1. Importing and preprocessing data
2. Building the model: Multinomial Naive Bayes
    - Model building
    - Model evaluation
## 1. Importing and Preprocessing Data

```python
import pandas as pd

# reading the training data
docs = pd.read_csv('smsspamcollection.csv', header = None, names=['Class','sms'])
docs.head()
```

```python
len(docs)
```

```python
ham_spam = docs.Class.value_counts()
ham_spam
```

```python
print("spam rate is about{0}%".format(
    round((ham_spam[1]/float(ham_spam[0]+ham_spam[1]))*100), 2))
```

```python
docs['label'] = docs.Class.map({'ham':0,'spam':1})
docs.head()
```

```python
# we can now drop the column 'Class'
docs = docs.drop('Class', axis=1)
docs.head()
```

```python
X = docs.sms
y = docs.label
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(stop_words='english')
```

```python
vect.fit(X_train)
```

```python
vect.vocabulary_
```

```python
len(vect.vocabulary_.keys())
```

```python
X_train_transformed = vect.transform(X_train)
X_test_transformed = vect.transform(X_test)
```

```python
print(type(X_train_transformed))
print(X_train_transformed)
```

## 2. Building and Evaluating the Model

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()

mnb.fit(X_train_transformed,y_train)

y_pred_class = mnb.predict(X_test_transformed)

y_pred_proba = mnb.predict_proba(X_test_transformed)
```

```python
mnb
```

### Model Evaluation

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred_class)
```

```python
metrics.confusion_matrix(y_test,y_pred_class)
```

```python
confusion = metrics.confusion_matrix(y_test, y_pred_class)
print(confusion)
TN = confusion[0,0]
FP = confusion[0,1]
FN = confusion[1,0]
TP = confusion[1,1]
```

```python
sensitivity = TP / float(TP+FN)
print('sensitivity',sensitivity)
```

```python
speecificity = TN / float(TN+FP)
print('speecificity',speecificity)
```

```python
precision = TP / float(TP+FP)
print('precision',precision)
print(metrics.precision_score(y_test,y_pred_class))
```

```python
print('Precision',precision)
print('Precision Score',metrics.precision_score(y_test,y_pred_class))
print('Recall Score',metrics.recall_score(y_test,y_pred_class))
print('F1 Score',metrics.f1_score(y_test,y_pred_class))
```

```python
y_pred_class
```

```python
y_pred_proba
```

```python
# creating an ROC curve
from sklearn.metrics import confusion_matrix as sk_confusion_matrix
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

false_postive_rate, true_postive_rate, thresholds = roc_curve(y_test, y_pred_proba[:,-1])
roc_auc = auc(false_postive_rate,true_postive_rate)
```

```python
# area undar the curve
print(roc_auc)
```

```python
# matrix of thersholds, tprm fpr
pd.DataFrame({'Threshold':thresholds,
            'TPR':true_postive_rate,
             'FPR':false_postive_rate
             })
```

```python
%matplotlib inline
plt.ylabel("True Positive Rate")
plt.xlabel('False Positive Rate')
plt.title('ROC')
plt.plot(false_postive_rate,true_postive_rate)
plt.show()
```

## 3. SMS Spam Classifier: Bernoulli Naive Bayes

```python
vect.vocabulary_
```

```python
vect.get_feature_names_out()
```

```python
# transform
X_train_transformed = vect.transform(X_train)
X_test_tranformed = vect.transform(X_test)
```

```python
from sklearn.naive_bayes import BernoulliNB

# instantiate bernoulli NB object
bnb = BernoulliNB()

# fit 
bnb.fit(X_train_transformed,y_train)

# predict class
y_pred_class = bnb.predict(X_test_tranformed)

# predict probability
y_pred_proba =bnb.predict_proba(X_test_tranformed)

# accuracy
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred_class)
```

```python
metrics.confusion_matrix(y_test,y_pred_class)
```

```python
confusion = metrics.confusion_matrix(y_test, y_pred_class)
print(confusion)
TN = confusion[0,0]
FP = confusion[0,1]
FN = confusion[1,0]
TP = confusion[1,1]
```

```python
sensitivity = TP / float(TP+FN)
print('sensitivity',sensitivity)
```

```python
speecificity = TN / float(TN+FP)
print('speecificity',speecificity)
```

```python
precision = TP / float(TP + FP)

print("precision",precision)
print(metrics.precision_score(y_test, y_pred_class))
```

```python
print('Precision',precision)
print('Precision Score',metrics.precision_score(y_test,y_pred_class))
print('Recall Score',metrics.recall_score(y_test,y_pred_class))
print('F1 Score',metrics.f1_score(y_test,y_pred_class))
```
